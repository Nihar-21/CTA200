```python
import numpy as np
import emcee

def log_prob(x, ivar):
    return -0.5 * np.sum(ivar * x ** 2)

ndim, nwalkers = 5, 100
ivar = 1. / np.random.rand(ndim)
p0 = np.random.randn(nwalkers, ndim)

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob, args=[ivar])
sampler.run_mcmc(p0, 10000)
```

```python
import numpy as np

#important that the first arguement of a probability function is the position of a single "walker"
#   which is an 'N' dimensional numpy array
def log_prob(x, mu, cov):
    diff = x - mu
    return -0.5 * np.dot(diff, np.linalg.solve(cov, diff))

#parameters set up in 5 dimensions
ndim = 5

np.random.seed(42)
means = np.random.rand(ndim)
print(means)

cov = 0.5 - np.random.rand(ndim**2).reshape((ndim, ndim))
print(cov)
cov = np.triu(cov)
print(cov)
print(cov.T)
cov += cov.T - np.diag(cov.diagonal())
cov = np.dot(cov, cov)
```

```python
nwalkers = 32
p0 = np.random.rand(nwalkers, ndim)
print(p0)
```

```python
#The main interface provided by emcee is th EnsebleSampler  object
import emcee
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob, args=[means, cov])

#Setting up the samplers with the argss we are saying that the probability function should be called as:
log_prob(p0[0], means, cov)   # ???? How is p0 sent in the line above???
#If we did'nt provide any args paramenter, tehn teh calling sequence couldhave to be: 
#   log_prob(p0[0]) instead
```

```python
# Its a good idea to run "burn-in" steaps to let the walkers explore the parameter space- get settled into the maximum of the density
state = sampler.run_mcmc(p0, 100)
sampler.reset()
```

```python
#EnsembleSampler.run_mcmc()
sampler.run_mcmc(state, 10000);
```

```python
import matplotlib.pyplot as plt

samples = sampler.get_chain(flat=True)   #Samples accessed (will return array with shape (10 000, 32, 5))
#You can make histograms of these samples to get an estimatet of the desnity that you were sampling
plt.hist(samples[:, 0], 100, color="k", histtype="step")
plt.xlabel(r"$\theta_1$")
plt.ylabel(r"$p(\theta_1)$")
plt.gca().set_yticks([]);
```

```python
#Another good test to see whether or not the sampling went well --> check the mean acceptance fraction of the ensemble
print(
    "Mean acceptance fraction: {0:.3f}".format(
        np.mean(sampler.acceptance_fraction)
    )
)
```

```python
#Also good to check the integrated autocorrelation time
print(
    "Mean autocorrelation time: {0:.3f} steps".format(
        np.mean(sampler.get_autocorr_time())
    )
)
```

                                          FITTING A MODEL TO DATA 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

# Choose the "true" parameters.
m_true = -0.9594
b_true = 4.294
f_true = 0.534

# Generate some synthetic data from the model.
N = 50
x = np.sort(10 * np.random.rand(N))
print(x)
yerr = 0.1 + 0.5 * np.random.rand(N)
print(yerr)
y = m_true * x + b_true
y += np.abs(f_true * y) * np.random.randn(N)
y += yerr * np.random.randn(N)

plt.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
x0 = np.linspace(0, 10, 500)
plt.plot(x0, m_true * x0 + b_true, "k", alpha=0.3, lw=3)
plt.xlim(0, 10)
plt.xlabel("x")
plt.ylabel("y");


#^^^^^HOW is the real data being modified? and how is emcee used to correct this?
#Difference between plotting (x,y, yerr) and (x0, m_true*x0 + b_true)?
```

```python
#^^^^Thick grey line = true model; underestimated uncertainities 
#Standard way to fit a l ine to data is Linear Least Squares 
#   https://www.youtube.com/watch?v=pKAPgUb4vL8

A = np.vander(x, 2)
# print(A)
C = np.diag(yerr * yerr)
# print(C)
ATA = np.dot(A.T, A / (yerr**2)[:, None])
print(ATA)
cov = np.linalg.inv(ATA)
print(cov)
w = np.linalg.solve(ATA, np.dot(A.T, y / yerr**2))
print("Least-squares estimates:")
print("m = {0:.3f} ± {1:.3f}".format(w[0], np.sqrt(cov[0, 0])))
print("b = {0:.3f} ± {1:.3f}".format(w[1], np.sqrt(cov[1, 1])))

plt.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0)
plt.plot(x0, m_true * x0 + b_true, "k", alpha=0.3, lw=3, label="truth")
plt.plot(x0, np.dot(np.vander(x0, 2), w), "--k", label="LS")
plt.legend(fontsize=14)
plt.xlim(0, 10)
plt.xlabel("x")
plt.ylabel("y");
```
