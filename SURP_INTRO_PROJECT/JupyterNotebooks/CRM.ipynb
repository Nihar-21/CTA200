```python
#***(2)**********PLOTTING a_vals and b_vals as a function of eb************#

#****************EXPIT FUNCTION**********************#
from scipy.optimize import curve_fit
from autograd.scipy.special import expit

#expit(x) = 1/(1+exp(-x))
def expit_reg(x,a,b):
    beta = a + b*x
    return expit(beta)

def ac_func(x,a,b,c):
    y = a + b*x + c*x**2
    return y

#****************************PLOTTING*****************#
fig = plt.figure(num=None, dpi=80, figsize = (6,15), facecolor='w', edgecolor='k')
fig.canvas.set_window_title('Window Title')

p0 = [-30,10]
avals , bvals = [], []

for i in range(0,len(len_eb),1):
    popt_p, pcov_p = curve_fit(expit_reg, ap_seg[i],p_seg[i], p0=p0, maxfev=5000)
    a_p,b_p = popt_p
    avals.append(a_p)
    bvals.append(b_p)

ax1 = fig.add_subplot(211)
# ax2 = fig.add_subplot(212)

p1 = [2,4,-2]
ac = -(np.array(avals)/np.array(bvals))
print(ac)


popt_ac, pcov_ac = curve_fit(ac_func,ebs, ac,p0=p1)
a_ac,b_ac,c_ac = popt_ac

print(popt_ac)

x_s_p = np.linspace(0.0, 0.7, 100)
y_s_p = ac_func(x_s_p, a_ac, b_ac, c_ac)


ax1.title.set_text('Critical Semi-major axis as a function of Binary Eccentricity') #Lambda_ = 100
ax1.set_xlabel('Binary Eccentricity (e$_b$)', fontsize = 12)
ax1.set_ylabel('a$_c$ (a$_b$)', fontsize = 12)
ax1.plot(ebs, ac, '+', color = 'green')
ax1.plot(x_s_p, y_s_p, '--', color = 'red')



# ax2.title.set_text('ebs vs. b_ values')
# ax2.set_xlabel('ebs', fontsize = 12)
# ax2.set_ylabel('b_values', fontsize = 12)
# ax2.plot(ebs, bvals, '+', color = 'green')

# plt.show()
# fig.savefig('Curve_fit(ac_func ;mu = {})(eb(0,25,1))(ap(0,25,1))_(lambda_=(100)_(NN)).png'.format(mu), dpi = 250)

```
