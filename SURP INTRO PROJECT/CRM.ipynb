```python

import numpy as np
import matplotlib.pyplot as plt 
import lifelines
from myfunctions import PlottingLL  
from myfunctions import CureModelLL
from myfunctions import SevenPlot


stime = np.load('stime_eb25_ap25_Mu0.50_Np30_logsp1.npy')
Ne, Na, Nm, Np = 25,25,1,30

#binary semi-major axis
ab = 1

#binary eccentricity
ebs = np.linspace(0.,0.7,Ne)
#print(ebs)

    
#test particle semi- major axis 
aps = ab*np.linspace(1.,5.,Na)


# for x in range(0,25,6):
#     print(ebs[x])
    
# for y in range(0,19,1):
#     print(round(aps[y],3))
    
#Binary reduced mass
mu = 0.500



# for y in range(0,19,1):
#     stimes.append(stime[0,y,:])
#     n = 30
    
#     app = [aps[y]]*n
#     ap_s.append(app)
    
# stimes = list(np.concatenate(stimes).flat)
# ap_s = list(np.concatenate(ap_s).flat)


p_s = []
lambda_s = []
ap_s = []

for y in range(0,19,1):
#     p_ = CureModelLL.CureModelLL(np.abs(ebs[0]),aps[y],mu,stime[0,y,:],Np)
#     print(p_)
    p_, lambda_ = CureModelLL.CureModelLL(np.abs(ebs[0]),aps[y],mu,stime[0,y,:],Np)
    lambda_s.append(lambda_)
    p_s.append(p_)
    ap_s.append(round(aps[y],3))
    



p_s = list(np.concatenate(p_s).flat)
lambda_s = list(np.concatenate(lambda_s).flat)

# print(ap_s)
# print(p_s)
# print(len(p_s))
# print(lambda_s)
# print(len(lambda_s))

fig = plt.figure()
gs = fig.add_gridspec(2, hspace=0)
axs = gs.subplots(sharex=True, sharey=False)
fig.suptitle('aps (1.0 - 4.0)')
axs[0].plot(ap_s, p_s, '+')
# axs[0].set_title('aps vs. p_')
axs[1].plot(ap_s, lambda_s, '+')
# axs[1].set_title('aps vs. lambda_')

axs[0].set(xlabel='aps', ylabel = 'p_')
axs[1].set(xlabel='aps', ylabel = 'lambda_')
fig.savefig('aps(1.0 - 4.0) vs. p_ & lambda_.png', dpi = 250)
    

# Hide x labels and tick labels for all but bottom plot.
# for ax in axs:
#     ax.label_outer()





```
